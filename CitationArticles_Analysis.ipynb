```python
#collection of citation information from all the json files and appending it to a dataframe and generating csv file.
import json
import pandas as pd
import numpy as np
from pandas import DataFrame
article_id = []
citingpaperId = []
citingdoi = []
citingtitle = []
citingvenue = []
citingyear = []
citingintent = []
citinginfluential = []
authornamelist= []
authoridlist = []
authorurllist = []
for i in range(115):
  f = open('/content/drive/My Drive/Latest_JSON_Files/Article' + str(i) + '.json')
  data = json.load(f)
  print("****** i",i)
  article_data = data['doi']
  for value in data['citations']:
    paperId = value['paperId']
    doi = value['doi']
    title = value['title']
    venue = value['venue']
    year = value['year']
    intent = value['intent']
    isInfluential = value['isInfluential']
    for i in value['authors']:
     citingpaperId.append(paperId)
     article_id.append(article_data)
     citingdoi.append(doi)
     citingtitle.append(title)
     citingvenue.append(venue)
     citingyear.append(year)
     citingintent.append(intent)
     citinginfluential.append(isInfluential) 
     authorname = i['name']
     authornamelist.append(authorname)
     authorid = i['authorId']
     authoridlist.append(authorid)
    #  authorurl= i['url']
    #  authorurllist.append(authorurl) 
a = {'article_id': article_id,'authorid': authoridlist,'authorname':authornamelist,'authorurl': authorurllist, 'paperId' : citingpaperId ,'doi' : citingdoi ,'title' : citingtitle ,
     'citing_venue' :citingvenue, 'year': citingyear,
     'intent' : citingintent,'isInfluential' : citinginfluential}
df = pd.DataFrame.from_dict(a, orient='index')
df1 = df.transpose()
df1['intent'] = df1['intent'].apply(lambda x : ' '.join(map(str, x)))
df1.to_csv('table3.csv', columns = a)
df1
```

```python
df = pd.read_csv("table3_upd.csv",encoding = "ISO-8859-1")
# df['Reference_paper_venue'] = df['Reference_paper_venue'].str.lower()
df1 = df[['citing_venue','article_id']]
df1.replace('', np.nan, inplace=True)
df1.drop_duplicates()
df1['citing_venue'] = df1['citing_venue'].str.lower()
print(df1.head())
```

```python
df2=df1.groupby('citing_venue')['article_id'].nunique().reset_index(name="count")
df2
df2.nlargest(5, ['count'])[["citing_venue","count"]]
df3 = df2.sort_values('count',ascending=False)
df3["Rank"] = df3["count"].rank(ascending=False)
df4 = df3.dropna(subset=['Rank'])
header = ["citing_venue", "Rank"]
df4.to_csv('citing_venue.csv', columns = header)
plotgraph = pd.read_csv("citing_venue.csv")
plotgraph = df2.nlargest(10, ['count'])[["citing_venue","count"]]
rank=pd.read_csv("citing_venue.csv")
top10rank = rank.head(10)
top10rank
# plotgraph['Rank'] = top10rank['Rank'].values
plotgraph.to_csv("citing_venue_f_l.csv")
```
