```python
#Plotting a graph for top 10 Insititutions
import pandas as pd
import matplotlib.pyplot as plt
plotgraph= pd.read_csv("top10Institutions_Vis.csv",encoding = "ISO-8859-1")
plotgraph.set_index("Insitutions",drop=True,inplace=True)
ax=plotgraph.plot.barh(color = 'blue')
ax.invert_yaxis()
venue=plotgraph['Count']
# plt.title("Top 10 Reference Articles Venues",fontsize=20)
plt.xlabel("Count",fontsize=14)
plt.ylabel("Insitutions",fontsize=14)
ax.legend().set_visible(False)
for i, v in enumerate(venue):
    ax.text(v, i, str(v),
            color = 'black', fontweight = 'bold')
plt.savefig('Top10Insitituions_graph.png',bbox_inches='tight',dpi=400)
plt.show()
# plotgraph= df.nlargest(10, ['count'])[["Reference_paper_venue","count"]]
# plotgraph
```
